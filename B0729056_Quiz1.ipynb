```python
from thinkdsp import Sinusoid
from thinkdsp import normalize, unbias
import numpy as np

class ShuWeiHsu_Signal(Sinusoid):
     def evaluate(self, ts):
        cycles = self.freq * ts + self.offset / np.pi / 2
        frac, _ = np.modf(cycles)
        ys = 2*(frac-1)*self.freq
        ys = normalize(unbias(ys), self.amp)
        return ys
```

```python
sawtooth = ShuWeiHsu_Signal(freq = 1000, amp = 10).make_wave(duration=1/200, framerate=40000)
sawtooth.plot()
```

```python
SawtoothSpectrum = sawtooth.make_spectrum()
SawtoothSpectrum.plot(high = 10000)
```
