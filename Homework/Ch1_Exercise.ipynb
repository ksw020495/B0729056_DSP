Exercise 2

```python
from thinkdsp import read_wave

wave = read_wave('07076039.wav')
wave.normalize()
wave.make_audio()
```

```python
wave.plot()
```

```python
segment = wave.segment(start = 2, duration = 0.5)
segment.plot()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot()
```

```python
spectrum1 = wave.make_spectrum()
spectrum1.high_pass(cutoff=5000, factor=0)
wave1 = spectrum1.make_wave()
spectrum1.plot()
wave1.make_audio()
```

聲音經過high_pass處理後變高，頻率<5000Hz的訊號被屏蔽

```python
spectrum2 = wave.make_spectrum()
spectrum2.low_pass(cutoff=5000, factor=0)
wave2 = spectrum2.make_wave()
spectrum2.plot()
wave2.make_audio()
```

聲音經過low_pass處理後變低，頻率>5000Hz的訊號被屏蔽

```python
spectrum3 = wave.make_spectrum()
spectrum3.band_stop(low_cutoff = 5000, high_cutoff = 10000, factor = 0)
wave3 = spectrum3.make_wave()
spectrum3.plot()
wave3.make_audio()
```

```python
聲音經過band_stop處理後，頻率介於5000至10000Hz的訊號被屏蔽
```

Exercise 3

```python
from thinkdsp import CosSignal, SinSignal

cos_sig1 = CosSignal(freq=440, amp=0.75, offset=0)
cos_sig2 = CosSignal(freq=880, amp=0.2, offset=0)
sin_sig1 = SinSignal(freq=1320, amp=0.5, offset=0)
mix = cos_sig1 + cos_sig2 + sin_sig1
mix.plot()
mix.make_wave().make_audio()
```

```python
mix.make_wave().make_spectrum().plot(high = 2000)
```

```python
sin_sig2 = SinSignal(freq=1760, amp=0.9, offset=0)
mix = mix + sin_sig2
mix.plot()
mix.make_wave().make_audio()
```

```python
mix.make_wave().make_spectrum().plot(high = 2000)
```

上方為加上去的波頻率為440的倍數的狀況

```python
sin_sig3 = SinSignal(freq=1000, amp=0.5, offset=0)
mix = mix + sin_sig3
mix.plot()
mix.make_wave().make_audio()
```

```python
mix.make_wave().make_spectrum().plot(high = 2000)
```

上方為加上去的波頻率為非440的倍數，聲音聽起來有疊不起來的感覺

```python
def stretch(wave, factor):
    wave.ts *= factor
    wave.framerate /= factor
```

```python
wave.plot()
```

```python
stretch(wave, 5)
wave.plot()
```

```python
時間軸間隔由原本的1變為5，而圖形看似不變，表示圖形被拉長了5倍
```
